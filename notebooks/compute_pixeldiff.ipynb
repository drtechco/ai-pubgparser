```python
import cv2 
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
img1 = cv2.imread("test_images/640_output_zh1_frame_0340.png", cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread("test_images/640_output_modzh1_frame_0340.png", cv2.IMREAD_GRAYSCALE)
height, width = img1.shape[:2]
crop_img1 = img1[30:300, 100-width:width-40]
crop_img2 = img2[30:300, 100-width:width-40]
ret1, thresh1 = cv2.threshold(crop_img1, 0, 128,cv2.THRESH_BINARY_INV | cv2.THRESH_OTSU)
ret1, thresh2 = cv2.threshold(crop_img2, 0, 128,cv2.THRESH_BINARY_INV | cv2.THRESH_OTSU)
diff_img = cv2.absdiff(thresh1, thresh2)
change_perc = np.count_nonzero(diff_img) / diff_img.size
print(change_perc)
stack = np.hstack((thresh1, thresh2, diff_img))
# stack = np.hstack((img1, img2))
plt.figure(figsize = (40,4))
plt.imshow(stack)
plt.show()
```

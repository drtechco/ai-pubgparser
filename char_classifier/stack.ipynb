```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
from torchvision import datasets, models, transforms
import matplotlib.pyplot as plt
import cv2
import numpy as np
from tqdm import tqdm
import os
%matplotlib inline
```

```python
def get_average(path_img: str):
    for file in os.listdir(path_img):
        imgs.append(os.path.join(path_img,file))
    acc = np.zeros((112, 112), dtype=np.float32)
    for _img in imgs:
        img = cv2.imread(_img)
        img = cv2.resize(img, (112,112))
        img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        # print(img.shape)
        acc += img
    acc = (acc / len(imgs)).astype(np.uint8)
    return acc
```

```python
imgs = []
path_img = "/home/hbdesk/labelstudio_convert/bbox_chars/O"
path_img2 = "/home/hbdesk/labelstudio_convert/bbox_chars/0"

acc = get_average(path_img)
acc2 = get_average(path_img2)
stack = np.hstack([acc, acc2])
plt.imshow(stack)
plt.show()
```

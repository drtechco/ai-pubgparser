```python
def update_results(task_id: int, results: dict):
    api_endpoint = f"/api/annotations/{task_id}"
    url = f"{base_url}{api_endpoint}"
    res = requests.patch(url, headers=headers, json=results)
    if res.status_code == 500:
        print(res.text)
    # pprint(res.json())
    # print("_________________-_______--_______")
    
    
def get_results(task_id: int):
    api_endpoint = f"/api/annotations/{task_id}"
    url = f"{base_url}{api_endpoint}"
    res = requests.get(url, headers=headers)
    pprint(res.json())

def get_all_annotations(project_id: int):
    api_endpoint = f"/api/projects/{project_id}/tasks/?page_size=-1"
    url = f"{base_url}{api_endpoint}"
    res = requests.get(url, headers=headers)
    if res.status_code == 200:
        return res.json()
    else:
        return None
```
